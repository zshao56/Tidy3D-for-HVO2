```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.integrate import trapezoid
import os
```

```python
file1_path = r'F:\OneDrive - UW-Madison\uwm\S4\data\output\blackbody_data-363K.txt'
data1 = np.loadtxt(file1_path, skiprows=1)

# Assuming data1 has two columns: x and y
x1, y1 = data1[:, 0], data1[:, 1]


file3_path = r'F:\OneDrive - UW-Madison\uwm\S4\data\output\divided.txt'
data3 = np.loadtxt(file3_path)

# Assuming data1 has two columns: x and y
x3 = data3

f_interpolate = interp1d(x1, y1, kind='linear', fill_value='extrapolate')
# Interpolate the first dataset at the x-coordinates of the second dataset
y1_interpolated = f_interpolate(x3)


integral_values_combined = []

integral1_values_combined = []

emissivity = []

for i in range(1, 10):  # 文件夹编号从1到9
    
    folder_path = "F:/OneDrive - UW-Madison/python/tidy 3d/HVO2emicam/data/"
    file2_path = os.path.join(folder_path, f"HVO0.0{i}.txt")
    data2 = np.loadtxt(file2_path, skiprows=1)
    


# Assuming data2 has multiple columns: x and multiple y columns
    x2, y2 = data2[:, 0], data2[:, 1]  # Transpose to have y-columns as rows

# Perform linear interpolation on the first dataset
    y2_interpolate = interp1d(x2, y2, kind='linear', fill_value='extrapolate')
    # Interpolate the first dataset at the x-coordinates of the second dataset
    y2_interpolated = y2_interpolate(x3)



    # Multiply the interpolated data with the current y-column of the second dataset
    y_combined = y1_interpolated * y2_interpolated

    y = y1_interpolated

    # Integrate the combined data
    integral_value_combined = trapezoid(y_combined, x3)

    integral1_value_combined = trapezoid(y, x3)

    e = integral_value_combined / integral1_value_combined

    integral_values_combined.append(integral_value_combined)

    integral1_values_combined.append(integral1_value_combined)

    emissivity.append(e)


```

```python
plt.plot(emissivity, marker='o', linestyle='-', label='Emissivity')
plt.title('Emissivity ht')
plt.xlabel('wvl')
plt.ylabel('Emissivity Value')
plt.legend()
plt.show()
```
