```python
import os
import matplotlib.pyplot as plt

# 文件夹路径
folder_path = "F:/OneDrive - UW-Madison/python/tidy 3d/HVO2emicam/data/"

# 初始化数据字典
data = {}

# 读取每个文件夹中的数据
for i in range(1, 10):  # 文件夹编号从1到9
    file_path = os.path.join(folder_path, f"HVO0.0{i}.txt")
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    x_values = []
    y_values = []
    for line in lines:
        x, y = map(float, line.split()) 
        x_values.append(x)
        y_values.append(y)
    
    data[f'{i}'] = {'x': x_values, 'y': y_values}

# 绘制曲线图
plt.figure(figsize=(10, 6))  # 设置图形大小
for folder, values in data.items():
    plt.plot(values['x'], values['y'], label=folder)

plt.xlabel('Wavelength (um)')
plt.ylabel('Absorption')
plt.title('Absorption with T')
plt.legend()  # 添加图例
plt.grid(True)
plt.show()
```

```python
import csv

# 文件夹路径
folder_path = "F:/OneDrive - UW-Madison/python/tidy 3d/HVO2emicam/data/"

# 待写入的CSV文件路径
output_csv_path = "F:/OneDrive - UW-Madison/python/tidy 3d/HVO2emicam/combined_data.csv"

# 打开CSV文件准备写入数据
with open(output_csv_path, 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    
    # 读取每个文件夹中的数据
    for i in range(1, 10):  # 文件夹编号从1到9
        file_path = os.path.join(folder_path, f"HVO0.0{i}.txt")
        with open(file_path, 'r') as file:
            lines = file.readlines()
        
        # 将数据写入CSV文件
        x_values = []
        y_values = []
        for line in lines:
            x, y = map(float, line.split())
            x_values.append(x)
            y_values.append(y)
        
        csv_writer.writerow(x_values)  # 写入 x 值
        csv_writer.writerow(y_values)  # 写入 y 值

print("数据已成功写入到 combined_data.csv 文件中。")
```
